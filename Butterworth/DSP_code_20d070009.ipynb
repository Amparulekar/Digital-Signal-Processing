```python
import cmath
import sympy
import numpy as np
import matplotlib.pyplot as plt
```

```python
poles = []
for i in range (19):
  p= 1j*1.0286*np.exp(1j*(2*i+1)*3.14159265/38)
  if (p.real<0):
    poles.append(p)
```

```python
def hana(s):
  denom=1
  for i in range(19):
    denom= denom*(s**2-poles[i]*0.7118*s+0.8478**2)
  return (0.73215*s)**19/denom
```

```python
w=np.linspace(0.01,2,100)
mag=[]
for i in range(w.shape[0]):
  mag.append(np.abs(hana(w[i]*1j)))
mag=np.array(mag)
plt.title('Magnitude vs frequency plot')
plt.xlabel('Omega')
plt.ylabel('Magnitude')
plt.plot(w,mag)
```

```python
w=np.linspace(0,5,20)
ph=[]
for i in range(w.shape[0]):
  ph.append(cmath.phase(hana(w[i]*1j)))
ph=np.array(ph)
plt.title('Phase vs frequency plot')
plt.xlabel('Omega')
plt.ylabel('Phase')
plt.plot(w,ph)
```

```python
denom=1
from sympy import expand, symbols
x=symbols('x')
for i in range(19):
  denom= denom*(x**2-poles[i]*0.7118*x+0.8478**2)
exp=expand(denom)
frac= (0.73215*x)**19/exp
print (frac)
```
